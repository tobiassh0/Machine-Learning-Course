### What is a Tensor?

It is a "Vector generalised to higher dimensions". Can have any number of dimensions 1,2,3,4 etc and they all represent something else. They are N dimensional arrays of base datatypes.

Each tensor has a datatype and a shape.
Datatypes include: float32, int32, string and others.
Shape: Dimension of data

#### Rank / Degree 
These are interchangeable terms which represent the dimension involved in the tensor. A string, or a singular number etc are all *rank 0* tensors. Rank 1 and 2 tensors are a list or a 2d list such as the below.

```python
rank1_tensor = tf.Variable(["Test"], tf.string)
rank2_tensor = tf.Variable([["test", "ok"], ["test", "yes"]], tf.string)
```

Determing the rank of a tensor we can call,

```python
tf.rank(rank2_tensor)
```

#### Shapes of Tensors
Can determine this by calling,

```python
rank2_tensor.shape
```

Which returns the [No. lists, No. elements per list]. 

Higher rank tensors will follow the same pattern [No. lists, No. lists inside of list, No. lists inside of list inside of list, ... , No. elements inside of last list].

Changing the shape of a tensor is useful for changing rank for a tensor containing the same amount of information and for doing algebra etc.

Can change the shape by using,

```python
tensor1 = tf.ones([1,2,3])  # tf.ones() creates a shape [1,2,3] tensor full of ones
tensor2 = tf.reshape(tensor1, [2,3,1])  # reshape existing data to shape [2,3,1]
tensor3 = tf.reshape(tensor2, [3, -1])  # -1 tells the tensor to calculate the size of the dimension in that place
                                        # this will reshape the tensor to [3,2]
                                                                             
# The numer of elements in the reshaped tensor MUST match the number in the original
```

Shape of [-1] infers the shape that it needs to be, i.e. using reshape [3,-1] shapes it in 3 lists with 2 elements.
